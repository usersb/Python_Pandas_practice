**Preliminaries**

```python
import pandas as pd
```

**Create a simulated dataset**

```python
#Create an example dataframe

data = {'Platoon' : ['A','A','A','A','A','A','B','B','B','B','B','C','C','C','C','C'],
        'Casualties' : [1,4,5,7,5,5,6,1,4,5,6,7,4,6,4,6]}

df = pd.DataFrame(data)
df
```

**Apply a function (Rolling Mean) to the dataframe by Group**

```python
df['rolling_mean'] = df.groupby('Platoon')['Casualties'].apply(lambda x: x.rolling(center = False,window = 2).mean())
df
```

```python
df['rolling_sum'] = df.groupby('Platoon')['Casualties'].apply(lambda x: x.rolling(center=False,window=2).sum())
df
```

```python
df.groupby('Platoon')['Casualties'].sum()
```
